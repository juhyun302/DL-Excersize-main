```python
# coding: utf-8
import os, sys, time
import numpy as np
import pickle
from dataset.mnist import load_mnist
from common.functions import sigmoid, softmax


def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, flatten=True, one_hot_label=False)
    return x_test, t_test

def init_network():
    # =======================================================
    # ğŸš¨ 1. ì—¬ê¸°! íŒŒì¼ì˜ 'ì „ì²´ ì ˆëŒ€ ê²½ë¡œ'ë¥¼ ì´ ë³€ìˆ˜ì— ë”± í•œë²ˆë§Œ ë„£ì–´ì¤˜!
    #    'ch02' í´ë”ë¥¼ í¬í•¨í•´ì„œ 'sample_weight.pkl'ê¹Œì§€ì˜ ì „ì²´ ê²½ë¡œì•¼.
    weight_file_path = r"C:\Users\EL19\Downloads\DL-Excersize-main\DL-Excersize-main\notebooks\DL3_20241006\ch02\sample_weight.pkl"
    # =======================================================

    print(">>> CWD:", os.getcwd()) 
    print(">>> íŒŒì¼ì„ ì°¾ì„ ìµœì¢… ì ˆëŒ€ ê²½ë¡œ:", weight_file_path)
    
    # ğŸ’¥ ì˜¤ë¡œì§€ ì´ ì ˆëŒ€ ê²½ë¡œë¡œë§Œ íŒŒì¼ì„ ì—´ê¸° ì‹œë„!
    with open(weight_file_path, 'rb') as f:
        network = pickle.load(f) # <--- ì´ì   ì§„ì§œ ì—´ë¦´ ê±°ì•¼!
    return network

def predict(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network ['b2'], network ['b3' ]
    
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)

    return y

# --- ì‹¤í–‰ ë¶€ë¶„ ---
x, t = get_data()
# os.chdir(PROJECT_ROOT_DIR) ê°™ì€ ì½”ë“œëŠ” ì´ì œ í•„ìš” ì—†ì–´!

# ğŸš¨ ì´ì „ì— PROJECT_ROOT_DIRì„ ì—†ì•´ìœ¼ë‹ˆ, ì´ ë¶€ë¶„ì€ ì‚­ì œí•´ì•¼ ì˜¤ë¥˜ê°€ ì•ˆ ë‚˜.
# network = init_network() # (ì›ë˜ëŒ€ë¡œë¼ë©´ ì´ ì½”ë“œë¥¼ ì¨ì•¼ í•´) 

# ë„¤ê°€ ë³µì‚¬í•´ì¤€ ì½”ë“œë¥¼ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•œë‹¤ë©´, ì•„ë˜ ì½”ë“œëŠ” ëƒ…ë‘¬ì•¼ê² ì§€.
# í•˜ì§€ë§Œ ìœ„ì—ì„œ PROJECT_ROOT_DIRì„ ì§€ì› ê¸° ë•Œë¬¸ì—, ì•„ë˜ ì½”ë“œëŠ” 'NameError'ê°€ ëœ° ê±°ì•¼.
# ì•„ë˜ ì½”ë“œë¥¼ ì§€ìš°ê³  ë‹¤ì‹œ ì‹¤í–‰í•´ì¤˜.

network = init_network()
accuracy_cnt = 0

for i in range(len(x)):
    y = predict(network, x[i]) 
    p = np.argmax(y)
    
    if p == t[i]:
        accuracy_cnt += 1 

print("\nì •í™•ë„(Accuracy): " + str(float(accuracy_cnt) / len(x) ))
```
